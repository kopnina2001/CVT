```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xscale('log')
ax.set_yscale('log')

diam = [0.400288, 0.131538, 0.128538, 0.0796924] #pol
errL2 = [5.718812e-03, 3.610559e-04, 4.396366e-03, 1.854612e-03] #pol

ord1 = diam
ord2 = [i ** 2 for i in diam]

ax.plot(diam, errL2, '-bo', label='$||u-u_N||_{L_2}$')
ax.plot(diam, ord1, '-r', label = '1-й порядок')
ax.plot(diam, ord2, '-g', label = '2-й порядок')
ax.legend()
plt.xlabel("Диаметр сетки")
plt.title("Для $u(x,y) = sin(x)\cdot sin(y)$ - сетка pol")
plt.show()
```
